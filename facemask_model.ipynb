```python
from facemask_model import *
import PIL
from utils import *
```

```python
dataset = create_dataset(path = 'data_extracted/')
train_dataloader, test_dataloader = create_dataloaders(dataset, train_ratio = 0.8, test_ratio = 0.2)
```

```python
model = FacemaskRecognizeModel()
```

```python
train(model, train_dataloader, n_epoch = 20, learning_rate = 0.01)
```

```python
# evaluate the model
n_predict, n_correct = evaluate(model, test_dataloader)
print('n_predict', n_predict)
print('n_correct', n_correct)
```

```python
# using model to predict an image
pathA = 'data_extracted/0/without_mask_3.jpg'
pathB = 'data_extracted/1/with_mask_2304.jpg'

imgA = PIL.Image.open(pathA)
imgB = PIL.Image.open(pathB)

predictA = predict(model, imgA)
predictB = predict(model, imgB)

print('predictA = {}, predictB = {}'.format(predictA, predictB))

show_side_by_side(pathA, pathB)
```
